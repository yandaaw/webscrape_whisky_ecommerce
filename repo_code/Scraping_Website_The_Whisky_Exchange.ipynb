```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://www.thewhiskyexchange.com/'

headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

productlinks = []
whiskylist = []

for x in range(1,6):
  r = requests.get(f'https://www.thewhiskyexchange.com/c/35/japanese-whisky?pg={x}#productlist-filter')
  soup = BeautifulSoup(r.content,'lxml')
  productlist = soup.find_all('li', class_='product-grid__item')
  for item in productlist:
    for link in item.find_all('a', href=True):
      productlinks.append(baseurl + link['href'])

for link in productlinks:
  r = requests.get(link, headers=headers)
  soup = BeautifulSoup(r.content,'lxml')

  name = soup.find('h1', class_='product-main__name').text.strip()
  content = soup.find('p', class_='product-main__data').text.strip()
  descriptions = soup.find('div', class_='product-main__description').text.strip()
  price = soup.find('p', class_='product-action__price').text.strip()
  try:
    #total_rating = soup.find('div', class_='review-overview').text.strip()
    total_rating = soup.find('span', class_='review-overview__rating star-rating star-rating--45').text.strip()
    total_reviews = soup.find('span', class_='review-overview__count').text.strip().replace(u'\xa0', ' ')
  except:
    total_rating = 'no rating'
    total_reviews = 'no reviews'

  whisky = {
      'name':name,
      'content':content,
      'rating':total_rating,
      'reviews':total_reviews,
      'descriptions':descriptions,
      'price':price
  }

  whiskylist.append(whisky)
```

```python
df = pd.DataFrame(whiskylist)
df
```
